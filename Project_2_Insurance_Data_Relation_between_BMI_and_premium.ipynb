# Deciding dependent and independant variables

```python
x = df[['bmi']]
```

```python
x
```

```python
y = df['charges']
```

```python
y
```

Train Test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3)
```

Select model ML

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

Training

```python
lm.fit(x_train,y_train)
```

Testing

```python
predictions = lm.predict(x_test)  # Testing
```

## Model Evaluation

Error calculation

error def.jpeg

```python
import numpy as np   # Evaluation 
import sklearn.metrics as metrics
print('MAE: {}'.format(metrics.mean_absolute_error(y_test, predictions)))
print('MSE: {}'.format(metrics.mean_squared_error(y_test, predictions)))
print('RMSE: {}'.format(np.sqrt(metrics.mean_squared_error(y_test, predictions))))
```

# Comapre Train, Test and Predicted Data 

```python
x = np.linspace(x.min(), x.max(), 100)
y = lm.predict(x.reshape(-1, 1))

import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure([
    go.Scatter(x=x_train.squeeze(), y=y_train, name='train', mode='markers'),
    go.Scatter(x=x_test.squeeze(), y=y_test, name='test', mode='markers'),
    go.Scatter(x=x, y=y, name='prediction')
])
fig.show()
```
